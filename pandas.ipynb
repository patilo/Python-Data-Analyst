*Pandas*  
Es la librería más importante de Python para la Ciencia de Datos y la Ingeniería de Datos, porque permite manipular tablas de datos de forma extremadamente sencilla.


**1. ¿Qué es Pandas y para qué sirve?**.  
Pandas es una librería de código abierto que proporciona estructuras de datos rápidas y flexibles. Su nombre viene de "Panel Data" (término econométrico).

*Sirve principalmente para:*  
Limpieza de datos: Manejar datos faltantes (NaN), eliminar duplicados o corregir formatos.

Transformación: Cambiar la estructura de tablas, pivotarlas o agruparlas.

Análisis: Realizar cálculos estadísticos rápidos sobre millones de filas.

Lectura/Escritura: Conectar con CSV, Excel, SQL, JSON y más.

Estructuras principales:
Series: Una columna (como una lista con esteroides).

DataFrame: Una tabla completa (como una hoja de Excel o una tabla de SQL).

Aquí aprendemos a cargar datos y darles un primer vistazo.

```python
import pandas as pd

# Crear un DataFrame desde un diccionario
datos = {
    'Producto': ['Leche', 'Tomate', 'Carne', 'Cloro'],
    'Precio': [1100, 1500, 9500, 1300],
    'Stock': [20, 50, 10, 5]
}

df = pd.DataFrame(datos)

# Visualización rápida
print(df.head(2))    # Ver las primeras 2 filas
print(df.info())     # Información de tipos de datos y nulos
print(df.describe().round(2)) # Estadísticas rápidas (media, max, min) y lo redondeamos a 2 decimas
```

Filtrado Lógico: Seleccionar solo lo que cumple una condición.

Manejo de Nulos: Decidir qué hacer con los datos vacíos.

```python
# 1. Filtrar productos con precio mayor a 2000
caros = df[df['Precio'] > 2000]

# 2. Crear una nueva columna calculada (Valor del inventario)
df['Total_Inventario'] = df['Precio'] * df['Stock']

# 3. Manejo de datos nulos (si existieran)
df.dropna()          # Elimina filas con nulos
df.fillna(0)         # Cambia nulos por un valor (ej: 0)

# 4. Agrupación (Group By) - Muy común en SQL
# ¿Cuánto stock total tenemos por categoría? (Si tuviéramos esa columna)
# total_por_cat = df.groupby('Categoria')['Stock'].sum()
```

Merges, Pivots y Optimización.  
Este es el nivel que necesitas para conectar diferentes fuentes de datos (ej. unir una tabla de ventas con una de clientes).

Merges (Joins): Unir tablas como lo harías en SQL (Left, Right, Inner, Outer).

Tablas Dinámicas (Pivot Tables): Resumir datos complejos.

Map y Apply: Aplicar funciones complejas a cada fila.
